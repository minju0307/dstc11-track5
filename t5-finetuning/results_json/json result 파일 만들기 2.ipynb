```python
import pandas as pd
import os
```

```python
clean_train=pd.read_csv("only_chatgpt_aug/predictions_1.csv")
```

```python
clean_train=clean_train["Generated Text"].tolist()
```

```python
len(clean_train)
```

```python
import re

clean_train_final=[]

for text in clean_train:
    try:
#         final_text=re.findall('S:(.+)<', text)[0]
        final_text=re.findall('(.+)<', text)[0]
        clean_train_final.append(final_text.strip())
    except:
        try:
            final_text=re.findall('>(.+)', text)[0]
            clean_train_final.append(final_text.strip())
        except:
            try:
                final_text=re.findall(':(.+)', text)[0]
                clean_train_final.append(final_text.strip())
            except:
                final_text=re.findall('(.+)', text)[0]
                clean_train_final.append(final_text.strip())
                print(final_text)
```

```python
clean_train_final
```

```python
import json

with open('baseline.ks.deberta-v3-base.json') as f:
    baseline=json.load(f)
```

```python
idx=0
for iii, chat in enumerate(baseline):
    if chat['target']==False:
        continue
    else:
        text=clean_train_final[idx]
        chat['response']=text
        idx+=1
```
