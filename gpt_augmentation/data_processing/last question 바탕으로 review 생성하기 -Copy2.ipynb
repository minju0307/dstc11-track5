## last question을 포함한 dialogue context 생성

```python
import os
text_dir='last_question_generation_train'
responses=os.listdir(text_dir)
```

```python
import re

p = re.compile(r'\d+')
responses=sorted(responses, key=lambda x : int(p.findall(x)[0]) )
```

```python
import json

with open("train_dialgoue_history_for_dataaug.txt") as f:
    train_dialogue_history=f.readlines()
```

```python
train_dialogue_history[4]
```

```python
## text 파일을 dict 형식으로 바꾸기 

train_dialogue_history_json=[]
error_idx=[]

for idx, d in enumerate(train_dialogue_history):
    if d.strip():
        try:
            d = d.replace("\n", "")
            d = d.replace("\'", "\"")
            d = re.sub("[a-zA-Z]\"[a-zA-Z]", "\'", d)
            dic=json.loads(d)
            train_dialogue_history_json.append([idx, dic]) ## orginal idx
        except:
            error_idx.append(idx)
```

```python
with open("train_last_question_aug.txt", "w") as fw:
    for idx, question in enumerate(responses):
        last_question=open(f"{text_dir}/last_question_train_{idx}.txt").read()
        fw.write("U: "+last_question+'\n')
```

```python
idx=0

for idx, dialogue_dict in train_dialogue_history_json:
    try:
        last_question=open(f"{text_dir}/last_question_train_{idx}.txt").read()
        dialogue_dict["dialogue context"].append("U: "+last_question)
        idx+=1
    except:
        continue
```

```python
with open("train_last_ques_aug_dialogue_history.txt", "w") as fw:
    for idx, dialogue in train_dialogue_history_json:
        fw.write(str(dialogue)+'\n')
    
```

## review 생성용 프롬프트 생성 

```python
## val dialogue history와 review knowledge 불러오기

val_dialogue_history=open("val_dialgoue_history.txt").readlines()
```

```python
data_dir='../dstc11-track5/data/val' ## val 

import json

with open(f'{data_dir}/labels.json', 'r', encoding='utf-8') as f:
    labels=json.load(f)
    
with open(f'{data_dir}/logs.json', 'r', encoding='utf-8') as f:
    logs=json.load(f)
```

```python
with open(f'../dstc11-track5/data/knowledge.json', 'r', encoding='utf-8') as f:
    knowledge=json.load(f)
```

```python
chats=[]
gold_response=[]

for idx, data in enumerate(logs):
    print(idx)
    one_chat={}
    
    ## knowledge
    if labels[idx]['target']==1:
        for k in labels[idx]['knowledge']:
            if k['doc_type']=='review':
                entity_name=knowledge[k['domain']][str(k["entity_id"])]['name']
                db=knowledge[k['domain']][str(k["entity_id"])]['reviews'][str(k["doc_id"])]['sentences'][str(k["sent_id"])]
                try:
                    one_chat[f'reviews of {entity_name}'].append(db)
                except:
                    one_chat[f'reviews of {entity_name}']=[]
                    one_chat[f'reviews of {entity_name}'].append(db)
                
            else:
                entity_name=knowledge[k['domain']][str(k["entity_id"])]['name']
                db=knowledge[k['domain']][str(k["entity_id"])]['faqs'][str(k["doc_id"])]['answer']
                try:
                    one_chat[f'FAQs of {entity_name}'].append(db)
                except:
                    one_chat[f'FAQs of {entity_name}']=[]
                    one_chat[f'FAQs of {entity_name}'].append(db)
    else:
            ## context
            one_chat["dialogue context"]=[]
    
    chats.append(one_chat)

    
```

```python
chats[0]
```

```python
with open("knowledge_val.txt", "w") as fw:
    for knowledge in chats:
        if str(knowledge)=="{'dialogue context': []}":
            fw.write('\n')
        else:
            fw.write(str(knowledge)+'\n')
```
